```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
#plt.rcParamsDefault['axes.unicode_minus']
%matplotlib inline
```

### 设置坐标轴

```python
figure = plt.figure(figsize=(3,5),edgecolor='r')
axes=figure.add_axes([0.1,0.1,0.7,0.8])# 栈图片的比例，为0~1之间的数
x = np.linspace(0,5)
y=x**2
line,=axes.plot(x,y)
axes.vlines(x=3,ymin=5,ymax=9,color='k')
axes.scatter(3,9,marker='*')
axes.spines['right'].set_color('none')
axes.spines['top'].set_color('none')
axes.xaxis.set_ticks_position('bottom')
axes.yaxis.set_ticks_position('left')
axes.spines['bottom'].set_position(('data',5))
axes.spines['left'].set_position(('data',2))
axes.set_yticks(np.arange(0,25,3),minor=False)# 是否设定小标签
#axes.set_xticklabels(list('ABCDEF'))
axes.set_xticklabels
axes.legend((line,),('$a\ simple\ line$',),loc='lower left')
```

### 绘制等高线

```python
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
n=256
x= np.linspace(-3,3,n)
y= np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)
plt.contourf(X,Y,f(X,Y),8,cmap=plt.cm.cool,alpha=0.75)
c=plt.contour(X,Y,f(X,Y),8,color='k',lw=1)
plt.clabel(c,inline=True,fontsize=10)
```

```python
from mpl_toolkits.mplot3d import Axes3D
mpl.rcParams['interactive']=True
fig = plt.figure()
axes = Axes3D(fig)
n= 256
x = np.linspace(-4,4,n)
y= np.linspace(-4,4,n)
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)
axes.plot_surface(X,Y,Z,cmap= plt.cm.hot)
axes.contourf(X,Y,Z,zdir='x',offset=-4)
#C = axes.coutour(X,Y,Z,color= 'k',lw = 2)
#axes.clabel(C, inline= True)
```

### 分格显示

```python
fig = plt.figure()
ax1= plt.subplot2grid((3,3),(0,0),colspan=3,rowspan=1)
ax2 = plt.subplot2grid((3,3),(1,0),colspan=2,rowspan=1)
ax3 = plt.subplot2grid((3,3),(1,2),colspan = 1, rowspan = 2)
ax4 = plt.subplot2grid((3,3),(2,0),colspan = 2,rowspan = 1)
plt.tight_layout(pad=0)
```

### 主次坐标轴

```python
x= np.linspace(-np.pi,np.pi)
y= np.exp(x)
z = np.sin(x)
fig, ax = plt.subplots()
ax2 = ax.twinx()
ax.plot(x,y)
ax2.plot(x,z)
```

### animation 动画

```python
from matplotlib import animation
fig, ax = plt.subplots() # Create a figure with a set of subplots already made
#plt.subplot() # Return a subplot axes positioned by the given grid definition.
x = np.linspace(0,3*np.pi,512)
line, = ax.plot(x, np.sin(x))

def anima(i):
    line.set_ydata(np.sin(x+i/100))
    return line,
def init():
    line.set_ydata(np.sin(x))
    return line,
ani = animation.FuncAnimation(fig = fig, func=anima , frames = 100, init_func = init , 
                              interval = 20,
                             blit = False)
plt.show()
```
