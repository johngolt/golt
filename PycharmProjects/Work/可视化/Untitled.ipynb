```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
sns.set(color_codes=True)
```

```python
titanic = sns.load_dataset("titanic")
sns.catplot(x="sex", y="survived", kind="bar", data=titanic);
```

```python
sns.catplot(x="sex", y="fare", kind="point", data=titanic,
            estimator=np.median)
```

```python
sns.catplot(x="sex", y="fare", kind="point", data=titanic,estimator=np.sum)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(alpha=0.3, bins=40, density = True,edgecolor='b')

plt.hist(x1, **kwargs);
plt.hist(x2, **kwargs);
plt.hist(x3, **kwargs);
```
