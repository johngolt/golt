#### balance_table

```python
balance = pd.read_csv('user_balance_table.csv')
balance['report_date'] = pd.to_datetime(balance['report_date'],
                                        format='%Y%m%d')
balance = balance.set_index('report_date')
```

```python
balance.isnull().sum()/balance.shape[0]
```

```python
# 剔除特征
balance = balance.drop(['category1','category2', 'category3', 
                       'category4'],axis=1)
```

```python
balance['user_id'].nunique(),df.shape
```

```python
# 第一天的用户数
balance.loc['2013-07-01','user_id'].nunique()
```

```python
# 每日流水的用户数
user_num = balance.groupby(by='report_date')['user_id'].nunique()
user_num.plot()
```

```python
balance.head()
```

```python
consume = balance.groupby('report_date')['consume_amt'].sum()
```

```python
pr = balance.groupby('report_date')['total_purchase_amt',
                             'total_redeem_amt'].sum()
```

```python
consume.plot() # 赎回中消费的曲线
```

```python
consume.nlargest() #双十一，情人节，跟购物相关的节日会用产生影响。
```

```python
tocard = pr['total_redeem_amt'] - consume
tocard.plot() #赎回中转移到银行卡的其他的。
```
